# Orbital Relationships Model

Predicting the orbit of Earth based on the orbits of Mercury, Venus and Mars.

```python
import numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow
```

```python
me_df=pd.read_csv('ephemeris_mercury.csv', index_col=False)
ve_df=pd.read_csv('ephemeris_venus.csv', index_col=False)
ea_df=pd.read_csv('ephemeris_earth.csv', index_col=False)
ma_df=pd.read_csv('ephemeris_mars.csv', index_col=False)
```

```python
for df in [me_df, ve_df, ea_df, ma_df]:
    df.drop(columns=['JDTDB', 'LT', 'RG', 'RR'], inplace=True)
    if df is not me_df:
        df.drop(columns=['Date'], inplace=True)
```
