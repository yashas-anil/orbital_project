## Prototype Orbital Model

Predicting the velocity of Mars (VX,VY,VZ) based on its position coordinates (X,Y,Z)

```python
import numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow
```

```python
eph_df=pd.read_csv('ephemeris.csv', index_col=False)
```

```python
eph_df.head()
```

```python
X=eph_df[['X','Y','Z']].values
y=eph_df[['VX','VY','VZ']].values
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
len(y_test)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3))
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=50)
```

```python
test_pred=model.predict(X_test)
test_pred
test_pred=pd.DataFrame(test_pred.reshape(5480,3))
# note you may need to adjust the reshape parameter based on the size of the training set
pred_df=pd.DataFrame(y_test)
pred_df=pd.concat([pred_df,test_pred],axis=1)
pred_df.columns=['True VX','True VY','True VZ','Predicted VX','Predicted VY','Predicted VZ']
pred_df.head()
```

```python
model.evaluate(X_test, y_test)
```
